##### Blank Pipeline

```python

from PIL import Image

image = Image.open('spacy_blank_pipeline.jpg')
image
```

```python
nlp=spacy.blank("en")

doc=nlp("Captain america ate 1005$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token)
```

```python
nlp=spacy.blank("en")

doc=nlp("Captain america ate 1005$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token," | ",token.pos_," | ",token.lemma_)
```

#### Above one is blank pipe line

```python
nlp.pipe_names
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
nlp.pipeline
```

```python
nlp.pipe_names
```

```python
doc=nlp("Captain america ate 1005$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token," | ",token.pos_," | ",token.lemma_)
```

```python
doc=nlp("Tesla Inc is going to acquire twitter for $45 billion")

for ent in doc.ents: #This doc property is used for the named entities in the document. If the entity recognizer has been applied, this property will return a tuple of named entity span objects.
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```

```python
# For a better way 
from spacy import displacy

displacy.render(doc,style='ent')
```

### Loaded Pipeline

```python
from PIL import Image

image = Image.open('spacy_loaded_pipeline.jpg')
image
```

```python
nlp=spacy.blank("en")

doc=nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:  
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```

It provide nothing 

```python
source_nlp=spacy.load("en_core_web_sm")

nlp=spacy.blank("en")

nlp.add_pipe('ner',source=source_nlp)
nlp.pipe_names
```

```python
doc=nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:  
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
```
